```python
import requests
import re
import csv
```

```python
response = requests.get('https://library.cfa.harvard.edu/journals')
journalpage = response.text
```

```python
string = journalpage.replace("\r", "").strip()
cleanstring = re.sub('\s+', ' ', string)
```

```python
m = re.findall(r'\<a href=\"(.*?)\"\>(.*?)\</a\>(.*?)\</td\>', cleanstring)
```

```python
journals = open("journals.csv",'w', newline='', encoding='utf-8-sig')

wr = csv.writer(journals, dialect='excel', quoting=csv.QUOTE_ALL)

print(m)
```

```python
for i in m:
    print (i[1])
    wr.writerow([i[0]]+[i[1]]+[i[2]])
journals.close()
```
